```python
#importing libraries
import PIL.Image
from tkinter import *
from PIL import Image,ImageTk

#Canvas
root = Tk()
canvas = Canvas(root, width=1000, height=800,bg="white")
canvas.pack()
#Setting default colour
colour = "brown"
pts = []
position_index= 1
root.title('SketchPad')
rct=[]
ln=[]
ply=[]
ovl=[]
fh=[]

#Function for clicking the button(downclick)
def push(event):
    global pts, position_index, colour
    if event.x <= 81:
        pts = []
        if event.y < 81:
            position_index = 1
        elif event.y < 161:
            position_index = 2
        elif event.y < 241:
            position_index = 3
        elif event.y < 321:
            position_index = 4
        elif event.y < 401:
            position_index = 5
            

    else:
        if position_index == 1 or position_index == 3 or position_index == 4 or position_index == 5:
            #Appending position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
        
        elif position_index == 2:
            #Appending initial position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
            if len(pts) >= 4:
                canvas.create_line(pts, fill=colour,tag="poly")
                


#Function while button is remain pressed and dragged
def drag(event):
    global pts, position_index, colour
    if event.x > 81:
        if position_index == 4:
            #Appending position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
            if len(pts) >= 4:
                #Creating shape and filing colour
                canvas.create_line(pts, fill=colour,tag="free_hand")
                
                
    

#Function after button is released(upclick)
def release(event):
    global pts, position_index, colour
    if event.x > 81:
        if position_index == 1:
            #Appending new position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
            #Creating shape and filing colour
            l1=canvas.create_line(pts, fill=colour,tag="line")
            ln.append(l1)
            pts = []
        elif position_index == 3:
            #Appending new position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
            #Creating shape and filing colour
            r1=canvas.create_rectangle(pts, fill=colour,tag="rect")
            rct.append(r1)
            pts = []
        elif position_index == 4:
            pts = []
            
        elif position_index == 5:
            #Appending new position x_axis and y_axis
            pts.append(event.x)
            pts.append(event.y)
            #Creating shape and filing colour
            o1=canvas.create_oval(pts, fill=colour,tag="oval")
            ovl.append(o1)
            pts = []


#Function to move the shapes
def move(event):
    #Widget of the event
    cmpnt=event.widget
    #Getting location
    loc_x, loc_y = cmpnt.winfo_x(), cmpnt.winfo_y()
    #Width and height 
    wd , ht =root.winfo_width(),root.winfo_height()
    mx ,my =cmpnt.winfo_width(),cmpnt.winfo_height()
    x_pos=(loc_x+event.x)-(15)
    y_pos=(loc_y+event.y)-int(my/2)
    if x_pos>=0 and y_pos>=0 and wd-abs(x_pos)>=0 and ht-abs(y_pos)>=0 and x_pos<=wd-5 and y_pos<=ht-5:
        cmpnt.place(x=x_pos,y=y_pos)   

#Function to save the image with button   
def save_as_png():
    # saving the postscipt image 
    canvas.postscript(file="file.ps", colormode='color')
    
    # using PIL to convert postscript to PNG 
    ps_image=PIL.Image.open("file.ps")
    ps_image.save('my_image.png')
    


#Open a saved image
def open_an_image(): 
    imgg = ImageTk.PhotoImage(Image.open("my_image.png"))
    canvas.create_image(100, 100, anchor=NW,image=imgg)
    mainloop()
    
    
#Function to clear canvas and show canvas   
def clear_all():
    canvas.delete("all")
    display_canvas()
    

def show_color(new_colour):
    #updating colours
    global colour
    colour = new_colour    
    

    
def display_canvas():
    
    #creating structue for icon for different modes
    canvas.create_line(0, 0, 81, 0)
    canvas.create_line(81, 0, 81, 391)
    canvas.create_line(0, 81, 81, 81)
    canvas.create_line(0, 161, 81, 161)
    canvas.create_line(0, 241, 81, 241)
    canvas.create_line(0, 321, 81, 321)

    ##Creating shape for icon of staright line
    canvas.create_line(16, 66, 66, 16)
    #Creating shape for icon of polygon
    canvas.create_line(16, 146, 46, 96, 66, 121)
    
    #Creating shape for icon of rectangle
    canvas.create_rectangle(16, 186, 66, 216)

    #Creating shape for icon of free hand
    canvas.create_line(41, 271, 41, 291)
    canvas.create_line(31, 281, 51, 281)
    canvas.create_line(36, 276, 46, 286)
    canvas.create_line(36, 286, 46, 276)
    
    #Creating shape for icon of oval
    canvas.create_oval(20,340,60,380)
    
    #structural line
    canvas.create_line(0, 390, 80, 390)
    
    id = canvas.create_rectangle((10,400,40,420),fill='black')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('black'))

    id = canvas.create_rectangle((10,430,40,450),fill='gray')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('gray'))

    id = canvas.create_rectangle((10,460,40,480),fill='brown4')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('brown4'))

    id = canvas.create_rectangle((10,490,40,510),fill='red')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('red'))

    id = canvas.create_rectangle((10,520,40,540),fill='orange')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('orange'))

    id = canvas.create_rectangle((10,550,40,570),fill='yellow')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('yellow'))

    id = canvas.create_rectangle((10,580,40,600),fill='green')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('green'))

    id = canvas.create_rectangle((10,610,40,630),fill='blue')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('blue'))

    id = canvas.create_rectangle((10,640,40,660),fill='purple')
    canvas.tag_bind(id, '<Button-1>', lambda x: show_color('purple'))
    

def key_press(event):

    x_axis, y_axis = 0, 0
    if event.char == "a":
        x_axis = -10
    if event.char == "d":
        x_axis = 10
    if event.char == "w":
        y_axis = -10
    if event.char == "s":
        y_axis = 10
    
    #moving the drawn shapes
    canvas.move(rct, x_axis, y_axis)
    canvas.move(ln, x_axis, y_axis)
    canvas.move(ovl, x_axis, y_axis)
    canvas.move(fh, x_axis, y_axis)
    
    
#Function to create a circle
def create_circle(x, y, r, canvas): #center coordinates, radius
    x0 = x - r
    y0 = y - r
    x1 = x + r
    y1 = y + r
    #canvas.create_oval(x0, y0, x1, y1,fill=colour)

create_circle(100, 100, 20, canvas)


#Function to create a square
def create_square(x,canvas):
    x0 = 200
    y0 = 200
    x1 = x 
    y1 = x
    #canvas.create_rectangle(x0,y0,x1,y1,fill=colour)

create_square(300,canvas)


#calling function to display canvas
display_canvas()

#Creating menu and it's submenu
menubar = Menu(root)
root.config(menu = menubar)
submenu = Menu(menubar,tearoff=0)
submenu1 = Menu(menubar,tearoff=0)

menubar.add_cascade(label='File', menu=submenu)
submenu.add_command(label='Save', command=save_as_png)
submenu.add_command(label='Open/Load Saved Image', command=open_an_image)
submenu.add_command(label='Clear', command=clear_all)

menubar.add_cascade(label='Delete', menu=submenu1)
submenu1.add_command(label='Rectangle/Square', command=lambda:canvas.delete("rect"))
submenu1.add_command(label='Oval/Circle', command=lambda:canvas.delete("oval"))
submenu1.add_command(label='Line', command=lambda:canvas.delete("line"))
submenu1.add_command(label='Free Hand', command=lambda:canvas.delete("free_hand"))
submenu1.add_command(label='Polygon', command=lambda:canvas.delete("poly"))

#Creating a label
lbl1 = Label(root, text = "Sketch Pad",bg='yellow', font=('times', 30, 'italic'))
lbl1.place(x=500,y=0)
lbl2 = Label(root, text = "Shapes",bg='green', font=('times', 20, 'italic'))
lbl2.place(x=85,y=0)

#Binding functions 
root.bind("<Key>", key_press)
root.bind("<Button-1>", push)
root.bind("<B1-Motion>", drag)
root.bind("<ButtonRelease-1>", release)
root.bind('<B3-Motion>',move)
mainloop()





```
